```python
import pandas as pd

df = pd.read_csv('vehicle_sales_data.csv')
df.head()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.head()
```

```python
# Check
print(df.isnull().sum())

# Optionally: drop or fill missing values
# df = df.dropna()
# or
# df['column_name'] = df['column_name'].fillna('Unknown')
df.head()
```

```python
df = df.drop_duplicates()
df.head()
```

```python
df['vehicle_type'] = df['vehicle_type'].str.strip().str.lower().str.title()
df['region'] = df['region'].str.strip().str.title()
df['supplier'] = df['supplier'].str.strip().str.title()
df.head()
```

```python
date_columns = ['sales_date', 'production_start_date', 'production_end_date']
for col in date_columns:
    df[col] = pd.to_datetime(df[col], dayfirst=True)
df.head()
```

```python
df['sales_quantity'] = df['sales_quantity'].astype(int)
df['price_per_vehicle'] = df['price_per_vehicle'].astype(int)
df['revenue'] = df['revenue'].astype(int)
df['production_quantity'] = df['production_quantity'].astype(int)
df['inventory_levels'] = df['inventory_levels'].astype(int)
df['cost_per_vehicle'] = df['cost_per_vehicle'].astype(int)
df['market_share_(%)'] = df['market_share_(%)'].astype(int)
df['profit_margin_(%)'] = df['profit_margin_(%)'].astype(int)
df['customer_satisfaction_rating'] = df['customer_satisfaction_rating'].astype(float)
```

```python
df.to_csv('cleaned_vehicle_sales_data.csv', index=False)
files.download('cleaned_vehicle_sales_data.csv')
```
